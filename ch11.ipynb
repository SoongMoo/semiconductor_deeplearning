## 4. 중요한 데이터 추출하기

```python
#plasma를 기준으로 각각 정상과 당뇨가 어느 정도 비율로 분포하는지 살펴봅니다. 
plt.hist(x=[df.plasma[df.diabetes==0], df.plasma[df.diabetes==1]], bins=30, histtype='barstacked', label=['normal','diabetes'])
plt.legend()                                                              #두개의 그래프가 하나로 겹쳐지게
#Plasma(공복 혈당 농도)와 diabetes(당뇨병)의 수치가 거의 같게 나타난 것을 확인 할 수가 있다. 서로 상관관계가 있다는 것을 알 수 있다.
```

```python
#BMI를 기준으로 각각 정상과 당뇨가 어느 정도 비율로 분포하는지 살펴봅니다. 
plt.hist(x=[df.bmi[df.diabetes==0], df.bmi[df.diabetes==1]], bins=30, histtype='barstacked', label=['normal','diabetes'])
plt.legend()
# bmi(채지방지수)와 diabetes(당뇨병)의 수치가 거의 같게 나타난 것을 확인 할 수가 있다. 
# 서로 상관관계가 있다는 것을 알 수 있다. BMI가 높아질 경우 당뇨의 발병률도 함께 증가하는 추세를 볼 수 있음
```

## 5. 피마 인디언 당뇨병 예측 실행

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense

# pandas 라이브러리를 불러옵니다.
import pandas as pd

# 피마 인디언 당뇨병 데이터셋을 불러옵니다. 
df = pd.read_csv('./data/pima-indians-diabetes3.csv')
```

```python
# 세부 정보를 X로 지정합니다.
X = df.iloc[:,0:8]
# 당뇨병 여부를 Y로 지정합니다.
y = df.iloc[:,8]
```

아래 코드를 보면 은닉층이 하나 더 생긴걸 확인 할 수 있다.<br />


```python
# 모델을 설정합니다.
model = Sequential()
model.add(Dense(12, input_dim=8, activation='relu', name='Dense_1'))
model.add(Dense(8, activation='relu', name='Dense_2'))
model.add(Dense(1, activation='sigmoid',name='Dense_3'))
model.summary()
# 층과 층의 연결을 한눈에 볼 수 있게 해 주는 model.summary() 부분이 추가
# (none,12)로 되어 있는데 none 에는 몇 개의 샘플이 들어가는지가 출력이된다.
# 행(샘플)의 수, Batch_size를 5로 주었지만 샘플 수가 출력되지 않았다.
# Param 부분은 파라미터 수, 즉 총 가중치와 바이어스 수의 합을 나타냄
# 예를 들어 첫 번째 층의 경우 입력 값 8개가 층 안에서 12개의 노드로 분산되므로 가중치가 8×12 = 96개가 되고, 
# 각 노드에 바이어스가 한 개씩 있으니 전체 파라미터 수는 96 + 12 = 108이 됨
```

```python
# 모델을 컴파일합니다.
model.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])

# 모델을 실행합니다.
history=model.fit(X, y, epochs=100, batch_size=5)
```

```python
import numpy as np

# 새로운 샘플 데이터 (1행 8열)
new_data = np.array([[6, 148, 72, 35, 0, 33.6, 0.627, 50]])

# 예측 수행
prediction = model.predict(new_data)
print(prediction)
# 예측 결과 출력
print(f"예측 확률: {prediction[0][0]:.4f}")
print(f"예측 결과 (0: 당뇨 아님, 1: 당뇨): {int(prediction[0][0] > 0.5)}")
```
