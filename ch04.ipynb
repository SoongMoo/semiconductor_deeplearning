## 4. 파이썬 코딩으로 확인하는 최소 제곱

### 1. 환경 준비

```python
import numpy as np # 수학적 계산을 파이썬에서 할 수 있게 하는 라이브러리
```

### 2. 데이터 준비

```python
# 공부한 시간과 점수를 각각 x, y라는 이름의 넘파이 배열로 만듭니다.
x = np.array([2, 4, 6, 8])     # 공부 시간 numpy배열 만듦
y = np.array([81, 93, 91, 97]) # 공부 시간에 따른 점수를 numpy 배열 만듦
x, y
```

#### [x와 y의 평균값]

```python
#x의 평균값을 구합니다.
mx = np.mean(x) # 시간인  배열 x에 대해서 평균을 구한다.

#y의 평균값을 구합니다.
my = np.mean(y) # 점수인 배열 y에 대해서 평균을 구한다.
# 출력으로 확인합니다.
print("x의 평균값:", mx)
print("y의 평균값:", my)
```

#### [기울기 공식의 분모와 분자]

```python
# 기울기 공식의 분모 부분입니다. 
divisor = sum([(i - mx)**2 for i in x])  

# 기울기 공식의 분자 부분입니다.
def top(x, mx, y, my):
    d = 0
    for i in range(len(x)):
        d += (x[i] - mx) * (y[i] - my)
    return d
dividend = top(x, mx, y, my)

# 출력으로 확인합니다.
print("분모:", divisor)
print("분자:", dividend)
```

### 3. 기울기와 y 절편 구하기

```python
# 기울기 a를 구하는 공식입니다.
a = dividend / divisor

# y절편 b 를 구하는 공식입니다.
b = my - (mx*a)

# 출력으로 확인합니다.
print("기울기 a =", a)
print("y절편 b =", b)
```

## 6. 파이썬 코딩으로 확인하는 평균 제곱 오차

### 1. 환경 준비

```python
import numpy as np
```

### 2. 데이터 준비

```python
#가상의 기울기 a와 y 절편 b를 정합니다.
fake_a=3   # 가상의 기울기
fake_b=76 # 가상의 절편

#공부 시간 x와 성적 y의 넘파이 배열을 만듭니다.
x = np.array([2, 4, 6, 8]) # 위와 같이 공부 시간 배열
y = np.array([81, 93, 91, 97]) # 위와 같이 공부 시간에 따른 점수 배열
```

### 3. 평균 제곱 오차 구하기

```python
# y=ax + b에 가상의 a,b 값을 대입한 결과를 출력하는 함수입니다.
def predict(x):
    return fake_a * x + fake_b    #  y = ax + b에 사간을 대입하여 예측 점수 확인

# 예측 값이 들어갈 빈 리스트를 만듭니다.
predict_result = [] # 예측점수를 저장하기 위한 리스트

# 모든 x값을 한 번씩 대입하여 predict_result 리스트를 완성합니다.
for i in range(len(x)):
    predict_result.append(predict(x[i]))
    print("공부시간=%.f, 실제점수=%.f, 예측점수=%.f" % (x[i], y[i], predict(x[i])))
```

```python
# 평균 제곱 오차 함수를 각 y값에 대입하여 최종 값을 구하는 함수입니다.
n=len(x)  
def mse(y, y_pred):
    return (1/n) * sum((y - y_pred)**2)

# 평균 제곱 오차 값을 출력합니다.
print("평균 제곱 오차: " + str(mse(y,predict_result))) # 평균제곱 오차가 적을 수록 최적의 선형회귀선
```
