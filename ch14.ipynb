## 데이터의 확인과 검증셋

```python
import warnings
warnings.filterwarnings('ignore')
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from sklearn.model_selection import train_test_split
import pandas as pd

# 데이터를 입력합니다.
df = pd.read_csv('./data/wine.csv', header=None)

# 데이터를 미리 보겠습니다.
df
```

```python
# 와인의 속성을 X로 와인의 분류를 y로 저장합니다.
X = df.iloc[:,0:12]
y = df.iloc[:,12]
```

데이터만 이용해서 학습을 시키면 과적합 오류가 발생할 수 있어 과적합 오류가 발생하지 않게 학습이 끝난 모델에 테스트를 하는 것이 테스트셋이라면 최적의 학습 파라미터를 찾기 위해 학습과정에서 사용하는 것이 검증셋이다.<br />
<br />
그래서 검증셋은 학습셋에서 일부를 사용하게 되는데 validation_split이라는 옵션을 사용하면된다 0.25는 80%에서 25%이므로 0.8*0.25=0.2가 되므로 전체에서 20%를 검증셋으로 사용한다는 의미가 된다

```python
#학습셋과 테스트셋으로 나눕니다.
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, shuffle=True)

# 모델 구조를 설정합니다.
model = Sequential()
model.add(Dense(30,  input_dim=12, activation='relu'))
model.add(Dense(12, activation='relu'))
model.add(Dense(8, activation='relu'))
model.add(Dense(1, activation='sigmoid'))
model.summary()

#모델을 컴파일합니다.
model.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])

# 모델을 실행합니다.
history=model.fit(X_train, y_train, epochs=50, batch_size=500, validation_split=0.25) # 0.8 x 0.25 = 0.2
# 검증셋(validation_split=0.25) : 학습을 하면서 학습이 얼마나 잘되고 있는지 검증한는 것 
# 검증셋을 두면 val_accuracy와 val_loss가 추가적으로 출력되는 것을 알 수 있다.
# accuracy: 훈련 데이터셋에 대한 모델의 정확도입니다. 훈련 과정 동안 모델이 훈련 데이터에 얼마나 잘 맞는지, 즉 얼마나 정확한 예측을 하는지를 나타냅니다.
# loss: 훈련 데이터셋에 대한 손실 값입니다. 손실 함수는 모델의 예측과 실제 값 간의 차이를 측정하는데 사용됩니다. 
#       훈련 과정에서 모델이 얼마나 잘 학습하고 있는지, 오차가 얼마나 줄어드는지를 알 수 있습니다.
```

```python
# 테스트 결과를 출력합니다.
score=model.evaluate(X_test, y_test)
print('Test accuracy:', score[1])
```

## 2. 모델 업데이트하기

epochs=50으로 줘서 50번 반복했을 때 지속적으로 제대로 학습을 하고 있는지를 알아 봐야 한다.
만약 지속적으로 좋아지는 것이 아니라면 최적의 상태에서 학습을 멈춰야 할 것이다.
즉, 학습을 많이 반복한다고 해서 모델 성능이 지속적으로 좋아지는 것은 아니므로 이를 적절히 정해 주는 것이 중요하다.
그리고 그 모델을 저장했다가 필요한 경우에 다시 불러서 사용하면 좋을 것이다.

### 기본 코드 불러오기

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from tensorflow.keras.callbacks import ModelCheckpoint
from sklearn.model_selection import train_test_split

import os
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# 데이터를 입력합니다.
df = pd.read_csv('./data/wine.csv', header=None)

# 와인의 속성을 X로 와인의 분류를 y로 저장합니다.
X = df.iloc[:,0:12]
y = df.iloc[:,12]

#학습셋과 테스트셋으로 나눕니다.
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, shuffle=True)

# 모델 구조를 설정합니다.
model = Sequential()
model.add(Dense(30,  input_dim=12, activation='relu'))
model.add(Dense(12, activation='relu'))
model.add(Dense(8, activation='relu'))
model.add(Dense(1, activation='sigmoid'))
model.summary()

#모델을 컴파일합니다.
model.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])
```

### 모델의 저장 설정 및 실행

```python
# 모델 저장의 조건을 설정합니다.
modelpath="./data/model/all/{epoch:02d}-{val_accuracy:.4f}.keras" # 파일을 생성했을 때 저장하는 경로를 지정해준다.
checkpointer = ModelCheckpoint(filepath=modelpath, verbose=1) # ModelCheckpoint은 모델을 체크해서 저장할 수 있게 도와주는 함수

# 모델을 실행합니다. 
history=model.fit(X_train, y_train, epochs=50, batch_size=500, validation_split=0.25, verbose=0, callbacks=[checkpointer])
# callbacks=[checkpointer] : 위에서 만든 checkpointer를 실행할 때 사용
# 실행시키면 50개의 파일이 만들어지는 것을 알 수 있다.
```

```python
# 테스트 결과를 출력합니다.
score=model.evaluate(X_test, y_test)
print('Test accuracy:', score[1])
```

## 3. 그래프로 과적합 확인하기

```python
학습의 진행과정을 그래프로 나타내보자.
학습된 결과를 history에 저장 했으므로 history를 이용해서 그래프를 출력할 수 있다.
```

```python
# 그래프 확인을 위한 긴 학습 (컴퓨터 환경에 따라 시간이 다소 걸릴수 있습니다)
# 위에서는 50번만 확인 했는데 다시 2000번을 학습을 시켜서 2000번의 정보를 저장하게 한다.
history=model.fit(X_train, y_train, epochs=2000, batch_size=500, verbose=0, validation_split=0.25) 
```

```python
# history에 저장된 학습 결과를 확인해 보겠습니다. 
# history에 저장된 학습결과를 2000번의 학습 결과를 pandas의 DataFrame으로 가져온다.
hist_df=pd.DataFrame(history.history)
hist_df
```

```python
# DataFrame으로 가져온 정보에서 val_loss과 loss만 불러와서 그래프로 그리도록 하자.
# y_vloss에  검증셋의 오차를 저장합니다.
y_vloss=hist_df['val_loss']

# y_loss에 학습셋의 오차를 저장합니다.
y_loss=hist_df['loss']

# x 값을 지정하고 검증셋의 오차를 빨간색으로, 학습셋의 오차를 파란색으로 표시합니다.
x_len = np.arange(len(y_loss))
plt.plot(x_len, y_vloss, "o", c="red", markersize=2, label='Testset_loss')
plt.plot(x_len, y_loss, "o", c="blue", markersize=2, label='Trainset_loss')

plt.legend(loc='upper right')
plt.xlabel('epoch')
plt.ylabel('loss')
plt.show()
```

```python
학습셋(파란색 선)의 오차는 줄어드는데 반해 검증셋(빨간색 선)의 오차는 어느순간 부터 다시 상승하는 것을 알 수 있다.
그러다는 것은 검증셋에서 더 이상 학습의 효과가 나아지지 않을 대 학습을 멈춰야 한다는 것이다.
```

## 4. 학습의 자동 중단

### 기본 코드 불러오기

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from sklearn.model_selection import train_test_split
from tensorflow.keras.callbacks import ModelCheckpoint, EarlyStopping
import os
import pandas as pd

# 데이터를 입력합니다.
df = pd.read_csv('./data/wine.csv', header=None)

# 와인의 속성을 X로 와인의 분류를 y로 저장합니다.
X = df.iloc[:,0:12]
y = df.iloc[:,12]

#학습셋과 테스트셋으로 나눕니다.
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, shuffle=True)

# 모델 구조를 설정합니다.
model = Sequential()
model.add(Dense(30,  input_dim=12, activation='relu'))
model.add(Dense(12, activation='relu'))
model.add(Dense(8, activation='relu'))
model.add(Dense(1, activation='sigmoid'))
model.summary()

#모델을 컴파일합니다.
model.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])
```

### 학습의 자동 중단 및 최적화 모델 저장

```python
저장할 파일을 지정하기 전에 먼저 자동으로 중지할 시점을 만들어줘야 한다
그것이 EarlyStopping이다.
```

```python
# 학습이 언제 자동 중단 될지를 설정합니다.
# 학습을 자동으로 중지하는대 val_loss(검증셋)가 20번이 지속되도록 나아지지 않으면 학습을 멈춰라.
early_stopping_callback = EarlyStopping(monitor='val_loss', patience=20)

#최적화 모델이 저장될 폴더와 모델의 이름을 정합니다.
modelpath="./data/model/Ch14-4-bestmodel.keras"

# 최적화 모델을 업데이트하고 저장합니다.                                        # save_best_only는 마지막 최적인 것 하나만 저장
checkpointer = ModelCheckpoint(filepath=modelpath, monitor='val_loss', verbose=0, save_best_only=True)

#모델을 실행합니다.                                                 # callbacks에 early_stopping_callback을 추가해서 검증샛의 결과가 나아지지 않으면 멈추어라
history=model.fit(X_train, y_train, epochs=2000, batch_size=500, validation_split=0.25, verbose=1, callbacks=[early_stopping_callback,checkpointer]) 
```

```python
# 테스트 결과를 출력를 확인 해보자, 
score=model.evaluate(X_test, y_test)
print('Test accuracy:', score[1])
```

```python
검증셋의 결과는 0.9846이고 테스트셋의 결과는 0.983846127986908으로 비슷한 것을 알 수 있다.
```

# 재 학습

```python
from tensorflow.keras.models import load_model
from tensorflow.keras.callbacks import ModelCheckpoint
import pandas as pd

# 데이터를 입력합니다.
df = pd.read_csv('https://raw.githubusercontent.com/SoongMoo/data/refs/heads/main/wine.csv', header=None)

# 와인의 속성을 X로 와인의 분류를 y로 저장합니다.
X = df.iloc[:,0:12]
y = df.iloc[:,12]

# 저장된 모델 불러오기
model = load_model('./wine.keras')

# 모델 구조 확인
model.summary()

# 모델 저장 조건 다시 설정
modelpath = "./wine.keras"
checkpointer = ModelCheckpoint(filepath=modelpath, verbose=1, save_best_only=True)

# 이전에 나눈 학습 데이터를 그대로 재학습
history = model.fit(
    X_train, y_train,
    epochs=30,          # 추가 학습 횟수
    batch_size=500,
    validation_split=0.25,
    verbose=1,
    callbacks=[checkpointer]
)

# 재학습 후 테스트 데이터로 평가
score = model.evaluate(X_test, y_test)
print(f"\n테스트 정확도: {score[1]*100:.2f}%")
```
