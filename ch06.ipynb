## 5. 텐서플로에서 실행하는 로지스틱 회귀 모델

### 1.환경 준비

```python
import numpy as np
import matplotlib.pyplot as plt
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
```

### 2. 데이터 준비

```python
x = np.array([2, 4, 6, 8, 10, 12, 14]) # 공부 시간
y = np.array([0, 0, 0, 1, 1, 1, 1]) # 0은 불합격 1은 합격
```

### 3. 모델 실행

손실함수 인 그래프<br />
  사용

```python
model = Sequential() # 모델 생성
model.add(Dense(1, input_dim=1, activation='sigmoid'))
   # 결과가 한이므로 1 , 입력값이 하나이므로 input_dim는 1, 두개중에 하나만 가져오는 것이므로 활성화 함수는 sigmoid이다.

# 교차 엔트로피 오차 함수를 이용하기 위하여 'binary_crossentropy'로 설정합니다. 
model.compile(optimizer='sgd' ,loss='binary_crossentropy') 
# optimizer는 경사 하강법, 손실함수는 교차엔트로피오차함수인 binary_crossentropy 사용
model.fit(x, y, epochs=1000)
```

### 4. 모델 테스트

```python
#그래프로 확인해 봅니다.

plt.scatter(x, y)
plt.plot(x, model.predict(x),'r')
plt.show()
```

```python
import tensorflow as tf
print("현재 TensorFlow 버전:", tf.__version__)
# 임의의 학습 시간을 집어넣어 합격 예상 확률을 예측해 보겠습니다.

hour = 7
input_data = tf.constant([[hour]], dtype=tf.float32) 
prediction = model.predict(input_data)[0][0]  


print("%.f시간을 공부할 경우, 합격 예상 확률은 %.01f%%입니다" % (hour, prediction * 100))
```
